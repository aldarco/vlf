```python
# para la estructura octogonal
c1 = 3.641
c2 = 0.925
```

```python
Na = 100       # numero de velktas
Aa = 7071      #area de antena
d = 1.72       # diametro del alambre
rho = 1.72e-8  # resistividad del cobre
```

```python
fv_sav = np.logspace(1, 6, 100)
```

```python
# savnet
c1_savnet = 4.0; c2_savnet = 1.217
Aa_savnet = 9
Na_savnet = 6
Ra_savnet = 4*rho*Na_savnet*c1_savnet*Aa_savnet**0.5/(np.pi*d**2)
La_savnet = 2*1e-7 * Na_savnet**2*c1_savnet*Aa_savnet**(1/2)*(np.log((c1_savnet*Aa_savnet**0.5)/(Na_savnet**0.5 *d) -c2_savnet))
#Zv = []
#for f in fv:
Zv_savnet = abs(Ra_savnet + 1j*2*np.pi*fv_sav*La_savnet)
```

```python
fv = np.logspace(1, 6, 1000)
turns = [10,20,50,100]
colors = np.linspace(0.2,1,len(turns))

for k, Na in enumerate(turns):
    Ra = 4*rho*Na*c1*Aa**0.5/(np.pi*d**2)
    La = 2*1e-7 * Na**2*c1*Aa**(1/2)*(np.log((c1*Aa**0.5)/(Na**0.5 *d) -c2))
    #Zv = []
    #for f in fv:
    Zv = abs(Ra + 1j*2*np.pi*fv*La)
    
    #Zv.append(abs(Z))
    plt.plot(fv/1000, Zv, label=f"{Na} ", color="firebrick", alpha=colors[k])
    print(f"{Na} (La={La*1e3:.2f} mH, Ra={Ra*1e6:.2f} uOhm)")
plt.tick_params(axis="both", which="both",direction="in")#, pad=-22)
#plt.tick_params(axis="x",direction="in")#, pad=-15)

plt.plot(fv_sav/1e3, Zv_savnet, "ks", label="Z_savnet", markerfacecolor="none", markersize=3)
plt.title("Impedancia")
plt.yscale("log")
plt.legend(title="Nro. vueltas")
plt.xlabel("Freq. (kHz)")
plt.ylabel("Z")
```
